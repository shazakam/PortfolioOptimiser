```python
import yfinance as yf
data = yf.Ticker('2020.OL')
```

```python
year_av = []
df = data.history(period="10y")
df.index = df.index.tz_localize(None)

for end_date in list(data.incomestmt.loc['Diluted EPS'].index):
    start_date = end_date - pd.DateOffset(years=1)
    average = df.loc[start_date:end_date]['Close'].mean()
    year_av.append(float(average))
```

```python
year_av.reverse()
```
